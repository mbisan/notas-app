```python
import os
```

```python
def list_all_items_os_walk(start_path):
    all_items = []
    for root, dirs, files in os.walk(start_path):
        all_items.append(root) 
        
        for directory in dirs:
            full_path = os.path.join(root, directory)
            all_items.append(full_path)
            
        for file in files:
            full_path = os.path.join(root, file)
            all_items.append(full_path)
            
    return sorted(all_items)
```

```python
list_all_items_os_walk('./notas')
```

```python
NOTES_DIR = './notas'
```

```python
def get_files_and_directories(path):
    result = []

    for f in os.listdir(path):
        cur_path = os.path.join(path, f)
        path_url = cur_path.replace(os.path.abspath(NOTES_DIR), '')
        if os.path.isdir(cur_path):
            subdir = get_files_and_directories(cur_path)
        else:
            subdir = None

        result.append({
            'path': path_url,
            'display': f,
            'subdir': subdir
        })

    return sorted(result, key=lambda x: x['display'] if x['subdir'] is None else '_' + x['display'])

get_files_and_directories('/home/martin/portfolio/notas_app/notas')
```
